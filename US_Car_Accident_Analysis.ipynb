```python
import seaborn as sns  # Import the seaborn library

# Load the 'car_crashes' dataset
carcrash = sns.load_dataset('car_crashes')
carcrash
```

The dataset represents U.S. state-level data on car crashes and related factors. Each row corresponds to a state (identified by the abbrev column) and includes statistics per 10,000 licensed drivers. Key variables include total crashes (total), those related to speeding (speeding) and alcohol (alcohol), as well as counts of crashes not involving distraction (not_distracted) or drivers with previous violations (no_previous). Insurance-related data is also included, such as average insurance premiums (ins_premium) and insurance losses (ins_losses). States like South Carolina (SC), North Dakota (ND), and Montana (MT) show notably high crash totals, while others like the District of Columbia (DC) and Massachusetts (MA) have lower values. This dataset can be used to analyze correlations between behavioral factors, crash rates, and insurance costs across states

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
carcrash = sns.load_dataset('car_crashes')

# Display the first few rows of the dataset
print(carcrash.head())

# Set the style
sns.set(style="whitegrid")

# 1. Barplot of total crashes per state (top 10)
plt.figure(figsize=(12, 6))
top_states = carcrash.sort_values('total', ascending=False).head(10)
sns.barplot(x='abbrev', y='total', data=top_states, palette='Reds_d')
plt.title('Top 10 States by Total Car Crashes')
plt.xlabel('State')
plt.ylabel('Total Crashes')
plt.show()

# 2. Scatter plot: speed-related crashes vs. alcohol-related crashes
plt.figure(figsize=(8, 6))
sns.scatterplot(x='speeding', y='alcohol', data=carcrash, hue='abbrev', palette='tab20', legend=False)
plt.title('Speeding vs Alcohol-Related Crashes')
plt.xlabel('Speeding Crashes per 10k Drivers')
plt.ylabel('Alcohol-Related Crashes per 10k Drivers')
plt.show()

# 3. Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(carcrash.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Car Crash Factors')
plt.show()
```

The scatter plot titled *"Speeding vs Alcohol-Related Crashes"* presents the relationship between speeding and alcohol-related crashes per 10,000 drivers across various states. The data points reveal a generally positive correlationstates with higher rates of speeding crashes also tend to exhibit elevated rates of alcohol-related crashes. This pattern suggests that reckless driving behaviors may co-occur, potentially driven by similar underlying risk factors such as impaired judgment, lax enforcement, or cultural attitudes toward driving safety. The spread of data, however, indicates variability: some states with moderate speeding crash rates show relatively high alcohol-related crash rates, and vice versa. This variability highlights the need for state-specific strategies that address the nuances of risky driving behaviors rather than relying on one-size-fits-all interventions.


```python
# 3. Correlation heatmap (corrected)
plt.figure(figsize=(10, 8))

# Select only numeric columns
numeric_data = carcrash.select_dtypes(include='number')

sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Car Crash Factors')
plt.show()
```

The correlation heatmap provides a visual representation of the linear relationships between various car crash-related factors across U.S. states. Notably, there is a very strong positive correlation between total crashes and the number of drivers with no previous violations (`0.96`), suggesting that higher crash rates are not necessarily due to prior infractions. Alcohol involvement (`0.85`) and non-distracted driving (`0.83`) also show strong correlations with total crashes, indicating that even in the absence of distraction, alcohol remains a critical risk factor. Speeding shows a moderate correlation with total crashes (`0.61`). Interestingly, insurance-related variables such as premiums and losses exhibit weak or even negative correlations with crash-related factorsfor instance, `ins_premium` correlates negatively with total crashes (`-0.20`), possibly reflecting higher premiums in states with fewer crashes as a risk mitigation strategy. Overall, the heatmap reveals that behavioral factors, especially alcohol and speeding, play a more significant role in crash incidence than financial variables.
